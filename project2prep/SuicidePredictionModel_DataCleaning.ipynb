```python
import pandas as pd
import numpy as np
```

Adding corruption index and mental health to full dataset

```python
corr_df = pd.read_pickle('corruption_data.pkl')
combined_health = pd.read_pickle('combined_df.pkl')
mental_health = pd.read_pickle('mental_health.pkl')
```

```python
combined_corr_health = pd.merge(corr_df, combined_health, 
                                on=['Country', 'Year'])
```

```python
#testing for view of one country
combined_corr_health[combined_corr_health.Country == 'Afghanistan']
```

```python
df = pd.merge(combined_corr_health, mental_health, on=['Country', 'Year'])
```

```python
#full combined dataset, years 2000 - 2015
df.head(10)
```

```python
df.Population = pd.to_numeric(df.Population)
```

```python
sorted_df = df.sort_values(by=['Country', 'Year']).reset_index(drop=True)
```

```python
sorted_df = sorted_df.dropna(thresh=6)
```

```python
sorted_df.Alcohol.fillna(method='ffill', inplace=True)
```

```python
sorted_df = sorted_df.drop(columns=['Hepatitis B', 'Status', 'under-five deaths '])
```

```python
sorted_df.rename(columns={'Rank': 'Corruption'}, inplace=True)
```

```python
sorted_df = sorted_df[((sorted_df.Corruption - sorted_df.Corruption.mean()) / sorted_df.Corruption.std()).abs() < 3]
```

```python
cleaned_df = sorted_df.groupby(['Country']).fillna(sorted_df.mean())
```

Adding target variable 

```python
suicide_df = pd.read_csv('/Users/leesurkis/Documents/Project2files/Downloaded_datasets/SuicideRates.csv')
```

```python
suicide_df.rename(columns={'Unnamed: 0' : 'Country', 
                                    'Unnamed: 1' : 'Year', 
                                    'Crude suicide rates (per 100 000 population)' : 
                                    'Suicide rate (per 100,000 population)'}, inplace=True)
```

```python
suicide_df = suicide_df[['Country', 'Year', 'Suicide rate (per 100,000 population)' ]]
suicide_df = suicide_df.iloc[1:]
```

```python
suicide_df.Year = pd.to_numeric(suicide_df.Year)
suicide_df['Suicide rate (per 100,000 population)'] = pd.to_numeric(suicide_df['Suicide rate (per 100,000 population)'])
```

```python
suicide_df.info()
```

Combining suicide data with total cleaned dataset

```python
#removing years not included in suicide data
cleaned_df = cleaned_df[cleaned_df['Year'].isin([2000, 2005, 2010, 2015])]
cleaned_df
```

```python
#merging with suicide data
combined_df = pd.merge(cleaned_df, suicide_df, on=['Country', 'Year'])
```

```python
combined_df.to_pickle('suicide_combined_df.pkl')
```

```python
#creating list for updated population outliers for year 2000
updated_values = []
for i in combined_df.Population[combined_df['Year'].isin([2000])]:
    updated_values.append(10*i)
```

```python
#list of population outliers year 2000
old_values = list(combined_df.Population[combined_df['Year'].isin([2000])])
```

```python
#replacing outliers with updated values
cleaned_pop = combined_df.replace(old_values, updated_values)
```

```python
#replacing 0 values with mean per country
cleaned_pop = cleaned_pop.replace(0.000000, method='ffill')
# fillnulls = sorted_df.groupby(['Country']).fillna(sorted_df.mean())
```

Converting people to % of population

```python
#checking which columns are measured in 'people'
cleaned_pop.iloc[:, 20:30].head(1)
```

```python
cleaned_pop.iloc[:, 3:4].head()
```

```python
cleaned_pop['Developmental intellectual disability (people)'] = cleaned_pop['Developmental intellectual disability (people)'] / cleaned_pop['Population']
cleaned_pop['Autistic spectrum disorders (people)'] = cleaned_pop['Autistic spectrum disorders (people)'] / cleaned_pop['Population']
cleaned_pop['Other disorders (people)'] = cleaned_pop['Other disorders (people)'] / cleaned_pop['Population']
cleaned_pop['Conduct disorder (people)'] = cleaned_pop['Conduct disorder (people)'] / cleaned_pop['Population']
cleaned_pop['Attention-deficit/hyperactivity disorder (people)'] = cleaned_pop['Attention-deficit/hyperactivity disorder (people)'] / cleaned_pop['Population']
cleaned_pop['Schizophrenia (people)'] = cleaned_pop['Schizophrenia (people)'] / cleaned_pop['Population']
cleaned_pop['Eating disorders (people)'] = cleaned_pop['Eating disorders (people)'] / cleaned_pop['Population']
cleaned_pop['Bipolar disorder (people)'] = cleaned_pop['Bipolar disorder (people)'] / cleaned_pop['Population']
cleaned_pop['Anxiety disorders (people)'] = cleaned_pop['Anxiety disorders (people)'] / cleaned_pop['Population']
cleaned_pop['Depressive disorders (people)'] = cleaned_pop['Depressive disorders (people)'] / cleaned_pop['Population']
```

```python
total_clean = cleaned_pop.rename(columns={'Developmental intellectual disability (people)':
                   'Developmental intellectual disability (percentage of population)',
                   'Autistic spectrum disorders (people)':
                   'Autistic spectrum disorders (percentage of population)', 
                   'Other disorders (people)':
                   'Other disorders (percentage of population)',
                   'Conduct disorder (people)':
                   'Conduct disorder (percentage of population)',
                   'Attention-deficit/hyperactivity disorder (people)':
                   'Attention-deficit/hyperactivity disorder (percentage of population)', 
                   'Schizophrenia (people)':
                   'Schizophrenia (percentage of population)', 
                   'Eating disorders (people)':
                   'Eating disorders (percentage of population)', 
                   'Bipolar disorder (people)':
                   'Bipolar disorder (percentage of population)', 
                   'Anxiety disorders (people)':
                   'Anxiety disorders (percentage of population)',
                   'Depressive disorders (people)':
                   'Depressive disorders (percentage of population)'})
```
